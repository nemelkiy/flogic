```python
import numpy as np
import pandas as pd
import os
from pandas import read_csv

column_names = ['CO2', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'STFc', 'STFs']

data = read_csv('C:/Users/kolom/fuzzy_logic_example/notebooks/input.csv', header=None, delimiter=r"\s+", 
                names=column_names, nrows=50)
new_data_frame = data.drop(['ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B'], axis=1)
print(new_data_frame)
```

```python
crim = new_data_frame['CO2'].values.tolist()
lstat = new_data_frame['STFc'].values.tolist()
medv = new_data_frame['STFs'].values.tolist()
print(crim, "\n", lstat, "\n", medv, "\n")
```

```python
max_crim, max_lstat, max_medv = max(crim), max(lstat), max(medv)
one_procent_crim = max_crim/100
one_procent_lstat = max_lstat/100
one_procent_medv = max_crim/100
norm_crim = [round((item/one_procent_crim)*0.1, 2) for item in crim]
print(norm_crim)
print(max_crim, max_lstat, max_medv)
```

```python
norm_crim, norm_lstat, norm_medv = [round(item/max_crim, 2) for item in crim], [round(item/max_lstat, 2) for item in lstat], [round(item/max_medv, 2) for item in medv]
print(norm_crim, "\n \n", norm_lstat,"\n \n", norm_medv)
```

```python
d = {'CO2': norm_crim, 'STFc': norm_medv, 'STFs': norm_lstat}
df = pd.DataFrame(data=d)
df.to_excel('C:/Users/kolom/fuzzy_logic_example/notebooks/normilizes.xlsx', index=False)
```
